```python
#representate this integer in binary then in 11 packages
def seed_to_mnemonic(seed):
    #binary conversion
    seed_bin = bin(seed)[2:].zfill(128)  # fill with some 0 to have 128 bytes

    checksum = hashlib.sha256(seed.to_bytes(16, byteorder='big')).hexdigest()
    checksum_bin = bin(int(checksum, 16))[2:].zfill(256)[:4]  # Take the 4 bytes of the checksum

    seed_bin_with_checksum = seed_bin + checksum_bin

    #divide the seed + checksum in segment of 11 bytes
    chunks = [seed_bin_with_checksum[i:i+11] for i in range(0, len(seed_bin_with_checksum), 11)]

    #attibute each segment to a word of the BIP 39 list
    mnemonic = [bip39_words[int(chunk, 2)] for chunk in chunks]

    return ' '.join(mnemonic)
    print(mnemonic)

seed_to_mnemonic(generate_seed())
```

```python
#attribute a word to each package
def mnemonic_to_seed(mnemonic):
    words = mnemonic.split()
    if len(words) != 12:
        raise ValueError("The mnemonic has to content 12 words.")

        #convert each word in this index in the BIP39 list
    bits = ''.join([bin(bip39_words.index(word))[2:].zfill(11) for word in words])

    #The original seed without the checksum (so 128 first bytes)
    seed_bin = bits[:128]

    #Convert the bytes in integer
    seed = int(seed_bin, 2)

    return seed

#let's do the process with the functions that we impleted before
seed = generate_seed()
print(f"Seed : {seed}")

mnemonic = seed_to_mnemonic(seed)
print(f"Phrase mnémonique : {mnemonic}")

#To import a mnemonic
imported_seed = mnemonic_to_seed(mnemonic)
print(f"Seed importée : {imported_seed}")
```

```python
import hashlib
import hmac
import binascii

def hmac_sha512(key, data):
    return hmac.new(key, data, hashlib.sha512).digest()

# Fonction de dérivation de clé enfant (clé privée) à l'index N
def derive_child_key_at_index(parent_private_key, parent_chain_code, index):
    # Convertir les entrées en bytes si ce n'est pas déjà le cas
    if isinstance(parent_private_key, str):
        parent_private_key = binascii.unhexlify(parent_private_key)
    if isinstance(parent_chain_code, str):
        parent_chain_code = binascii.unhexlify(parent_chain_code)

    # Préparer l'entrée pour l'HMAC : clé privée + index (en big-endian sur 4 octets)
    index_bytes = index.to_bytes(4, byteorder='big')

    # Appliquer HMAC-SHA512 avec la clé : parent_chain_code, data : (0x00 + parent_private_key + index)
    data = b'\x00' + parent_private_key + index_bytes
    hmac_result = hmac_sha512(parent_chain_code, data)

    # Diviser la sortie HMAC en deux : les premiers 256 bits (32 octets) pour la nouvelle clé privée,
    # et les derniers 256 bits (32 octets) pour le chain code
    new_private_key = hmac_result[:32]
    new_chain_code = hmac_result[32:]

    return new_private_key, new_chain_code

# Fonction récursive pour générer une clé à un index N au niveau de dérivation M
def derive_key_at_level(parent_private_key, parent_chain_code, index_list):
    current_private_key = parent_private_key
    current_chain_code = parent_chain_code

    # On parcourt la liste des indices et on génère une clé à chaque niveau
    for index in index_list:
        current_private_key, current_chain_code = derive_child_key_at_index(current_private_key, current_chain_code, index)

    return current_private_key, current_chain_code

# Exemple d'utilisation pour générer une clé enfant à l'index N au niveau M
parent_private_key = "f1c620f43c00c6b75a4d1b3c8b3d05b9cde2fc9b7f3c6a1d8b44b6d0e021d1f9"  # Exemple de clé privée parent en hex
parent_chain_code = "6048ff3b5e7f0c6f276e898d98ff99423799fc3132bcb769ac320a1d4ba93dc3"  # Exemple de chain code parent en hex
index_list = [5, 10, 20]  # Exemple de niveaux : générer la clé enfant au niveau 3 (indices 5, 10, 20)

child_private_key, child_chain_code = derive_key_at_level(parent_private_key, parent_chain_code, index_list)
print(f"Child Private Key at level M: {binascii.hexlify(child_private_key).decode()}")
print(f"Child Chain Code at level M: {binascii.hexlify(child_chain_code).decode()}")
```
